## Open the FOIA document via Pandas

```python
from pathlib import Path
import pandas as pd

ROOT = Path('../..') 
DATA_PATH = ROOT / 'src' / 'data'

df = pd.read_excel(DATA_PATH / 'raw' / 'xlsx' / 'complaints.xlsx')
df
```

## Drop the two Unnamed: 11	& Unnamed: 12 columns 

```python
df.dtypes
```

```python
# Source - https://stackoverflow.com/a/18145399
# Posted by LondonRob, modified by community. See post 'Timeline' for change history
# Retrieved 2026-01-30, License - CC BY-SA 4.0

df = df.drop('Unnamed: 11', axis=1)
df = df.drop('Unnamed: 12', axis=1)
df
```

## I want the final csv to show the number of complaints about JH Baxter per year. 

### To do this, let's make a separate year column. 

```python
df['Date of Complaint Submission'] = pd.to_datetime(
    df['Date of Complaint Submission'],
    errors='coerce'
)

df['Year'] = df['Date of Complaint Submission'].dt.year
df
```

## Awesome, now it's time to make a separate df. This new df will have the year with the number of complaints per year. 

```python
df_total_complaints = (
    df['Year']
    .value_counts()
    .sort_index()
    .reset_index()
)

df_total_complaints.columns = ['Year', 'total_complaints']
df_total_complaints.to_csv(DATA_PATH / 'processed' / 'csv' / 'datawrapper-complaints.csv', index=False)

```
